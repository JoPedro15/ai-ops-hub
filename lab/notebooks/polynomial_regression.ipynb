```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import scatter
from sklearn.metrics import r2_score
```

```python
np.random.seed(2)
load_time = np.random.normal(3.0, 1.0, 1000)
load_time = np.maximum(load_time, 0.1)

purchase_amount = np.random.normal(50.0, 10.0, 1000) / load_time
purchase_amount = np.maximum(purchase_amount, 0.0)

scatter(load_time, purchase_amount)
```

```python
x = np.array(load_time)
y = np.array(purchase_amount)

p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c="red")
plt.show()
```

```python
r2 = r2_score(y, p4(x))
```
