```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import scatter
from scipy import stats
from sklearn.metrics import r2_score
```

```python
load_time = np.random.normal(5.0, 3.0, 1000)
load_time = np.maximum(load_time, 0.1)

purchases_amount = 100 - (load_time + np.random.normal(0, 1, 1000)) * 3
purchases_amount = np.maximum(purchases_amount, 0.0)

scatter(load_time, purchases_amount)
```

```python
slope, intercept, r_value, p_value, std_error = stats.linregress(
    load_time, purchases_amount
)
```

```python
def predict(x):
    return slope * x + intercept


fit_line = predict(load_time)
```

#### Get R-squared

Varies between 0 and 1

The closer we get to 1 the fitter our regression is.

```python
r2_score(purchases_amount, fit_line)
```

```python
plt.scatter(load_time, purchases_amount)
plt.plot(load_time, fit_line, c="red")
plt.xlabel("Tempo de Carregamento (s)")
plt.ylabel("Gasto (â‚¬)")
plt.show()
```
